```python
import numpy as np
import pandas as pd
dates = pd.date_range("20130101", periods=6)
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
print(df[df["A"] > 0])
df.at[dates[0], "A"] = 0  #set values by label
df.iat[0, 1] = 0 #set values by position
print()
print(df)
#df.dropna(how="any")   drop any row with missing data
#df.fillna(value=5)   fill missing values with provided
# isna() gets boolean
```

```python
import numpy as np
import pandas as pd
dates = pd.date_range("20130101", periods=6)
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
print(df.mean()) #mean of each row
print()
print(df.agg(lambda x: np.mean(x) * 5.6))
print()
print(df.transform(lambda x: x * 101.2))
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add species column
df['species'] = iris.target
df['species'] = df['species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})
print(df.head())
print()
print(df.info())
```
